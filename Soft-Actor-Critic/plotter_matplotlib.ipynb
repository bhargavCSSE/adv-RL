```python
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Import files and Generate plot data
plotdata_1 = pd.read_csv('data/SAC-LunarLander1000-rewards-train.csv', index_col=0)
plotdata_2 = pd.read_csv('data/SAC-LunarLander1000-rewards-test.csv', index_col=0)

n_episodes_train, n_trials_train = plotdata_1.shape
n_episodes_test, n_trials_test = plotdata_2.shape
train_x = [i+1 for i in range(n_episodes_train)]
test_x = [i+1 for i in range(n_episodes_test)]

mean_vals_1 = plotdata_1.mean(axis=1)
std_vals_1 = plotdata_1.std(axis=1)
mean_vals_2 = plotdata_2.mean(axis=1)
std_vals_2 = plotdata_2.std(axis=1)
```

```python
plt.figure()

plt.title("Environment: LunarLanderContinuous-v2, n_trials: 5")
plt.xlabel('Episodes')
plt.ylabel('Cumulative average reward')

plt.plot(train_x, mean_vals_1, label='SAC (training)')
plt.fill_between(train_x, mean_vals_1-std_vals_1, mean_vals_1+std_vals_1, alpha=0.5)

plt.legend(loc='lower right')

plt.savefig('plots/LunarLander1000-training')
plt.show()
```

```python
plt.figure()

plt.title("Environment: LunarLanderContinuous-v2, n_trials: 5")
plt.xlabel('Episodes')
plt.ylabel('Cumulative average reward')

plt.plot(test_x, mean_vals_2, label='SAC (testing)')
plt.fill_between(test_x, mean_vals_2-std_vals_2, mean_vals_2+std_vals_2, alpha=0.5)

plt.legend(loc='lower right')

plt.savefig('plots/LunarLander1000-test')
plt.show()
```
